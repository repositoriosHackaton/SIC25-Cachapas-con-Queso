````python
import time

set_de_ingredientes = set()

for i in range(790, 796):

    print(i)
    if (i % 15 == 0):
        with open("mi_archivo.txt", "a") as archivo:
            for ingrediente in set_de_ingredientes:
                archivo.write(ingrediente + "\n")
        set_de_ingredientes.clear()
        time.sleep(65)

    prompt = str(ingredientes_csv[i])
    response = chat.send_message(prompt)

    lista_de_ingredientes = eval(response.text.replace("python", "").replace("```", "").replace("```", "").strip())
    for ingrediente in lista_de_ingredientes:
        set_de_ingredientes.add(ingrediente)

with open("mi_archivo.txt", "a") as archivo:
    for ingrediente in set_de_ingredientes:
         archivo.write(ingrediente + "\n")

set_de_ingredientes.clear()
````
